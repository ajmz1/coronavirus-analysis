```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = r'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'

df = pd.read_csv(url)
```

```python
df.to_csv('full-data.csv', index=False)
```

```python
print(df.columns)
df.describe()
```

```python
# First and last date
dates = np.sort(df['date'].unique())
print(dates[0], dates[-1])
```

```python
df_usa = df[df.location == 'United States']
df_usa.head(10)
```

```python
import datetime

#columns = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'reproduction_rate']
columns = ['total_cases', 'total_deaths']

fig, ax = plt.subplots()
ax.minorticks_on()
x = pd.to_datetime(df_usa.date, format='%Y/%m/%d')

for idx, colName in enumerate(columns):
    ax.plot(x, df_usa[colName], lw=2.2, label=colName)


ax.set(yscale='log')



ax.grid(which='major', axis='both', color='k', linestyle='-')
ax.grid(True, which='minor', axis='y', color='k', linestyle=':')
plt.legend()
```
